```python
import pandas as pd
import numpy as np
import glob
```

```python
# Setting the main filepath and using glob to get the filepaths of all xlsx files in a folder
main_filepath = "C:/Users/TooFastDan/Documents/MD_PhD Application/Docs2be/AAMC Data/A-2"
excel_files = glob.glob(main_filepath + "/*.xlsx")
```

```python
a2_list = []
for xlsx in excel_files:
    a2 = pd.read_excel(xlsx)
    a2_year = a2.iloc[0,0]
    a2 = a2[a2.iloc[:, 1].notnull()].reset_index(drop=True)
    a2.columns = a2.iloc[0]
    a2 = a2.drop(0).reset_index(drop=True)
    a2["year"] = a2_year
    a2_list.append(a2)
a2 = pd.concat(a2_list).reset_index(drop=True)
a2.columns = [i.replace("\n", "") for i in a2.columns]
a2["Total Applicants from the Institution"] = pd.to_numeric(a2["Total Applicants from the Institution"])
a2
```
